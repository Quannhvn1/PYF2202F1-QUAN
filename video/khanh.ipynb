```python
from tkinter import *
from tkinter import ttk
from tkinter import messagebox

root = Tk()
root.title("Quản lý bài giảng của giaó viên")
root.geometry("900x500")

#Tạo treeview
my_tree = ttk.Treeview(root)

my_tree["columns"] = ("classes", "subject", "date", "time", "content")
my_tree.column("#0", width=0, stretch=NO)
my_tree.column("classes", anchor=CENTER, width=60)
my_tree.column("subject", anchor=CENTER, width=90)
my_tree.column("date", anchor=CENTER, width=90)
my_tree.column("time", anchor=CENTER, width=90)
my_tree.column("content", anchor=CENTER, width=200)

my_tree.heading("#0", text="Label", anchor=CENTER)
my_tree.heading("classes", text="Lớp học", anchor=CENTER)
my_tree.heading("subject", text="Môn dạy", anchor=CENTER)
my_tree.heading("date", text="Ngày dạy", anchor=CENTER)
my_tree.heading("time", text="Thời gian dạy", anchor=CENTER)
my_tree.heading("content", text="Nội dung dạy", anchor=CENTER)

my_tree.place(x=300, y=130, width=550)

#data 
data = [["10A", "Tin học", "1/4", "45 phút", "Chuyển đổi nhị phân"], ["12B", "Toán", "1/4", "90 phút", "Hàm số bậc cao"],
       ["12B", "Toán", "4/4", "90 phút", "Ôn tập"], ["10B", "Tin", "6/4", "45 phút", "Thông tin và dữ liệu"],
       ["10D", "Tin", "8/4", "45 phút", "Hệ diều hành"], ["12B", "Toán", "9/4", "90 phút", "Lượng giác"]]   

global count
count = 0
for record in data:
    my_tree.insert(parent="", index="end", iid=count, text="Parent", values=([record[0], record[1], record[2],record[3], record[4]]))
    count += 1

classes_lbl = Label(root, text="Quản lý bài giảng của giáo viên", font=("Arial", 25))
classes_lbl.place(x=200, y=30)
classes_lbl = Label(root, text="Lớp học")
classes_lbl.place(x=20, y=130)
subject_lbl = Label(root, text="Môn dạy")
subject_lbl.place(x=20, y=160)
date_lbl = Label(root, text="Ngày dạy")
date_lbl.place(x=20, y=190)
time_lbl = Label(root, text="Thời gian dạy")
time_lbl.place(x=20, y=220)
content_lbl = Label(root, text="Nội dung dạy")
content_lbl.place(x=20, y=250)

classes = Entry(root)
classes.place(x=100, y=130)
subject = Entry(root)
subject.place(x=100, y=160)
date = Entry(root)
date.place(x=100, y=190)
time = Entry(root)
time.place(x=100, y=220)
content = Entry(root)
content.place(x=100, y=250)
search = Entry(root)
search.place(x=120, y=400)

#add info
def add_record():
    global count
    global data
    input_data = [classes.get(), subject.get(), date.get(), time.get(), content.get()]
    data.append(input_data)
    if classes.get() == "" or subject.get() == "" or date.get() == "" or time.get() == "" or content.get() == "":
        messagebox.showinfo("Error", "Entry cannot be empty")
    else:
        my_tree.insert(parent="", index="end", iid=count, text="Parent", values=(classes.get(), subject.get(),\
                                                                              date.get(), time.get(), content.get()))
        count += 1
        messagebox.showinfo("Add information", "Information added sucessfully!")
        clear_entry()     
    print(data)      

#clear entry
def clear_entry():
    classes.delete(0,END)
    subject.delete(0,END)
    date.delete(0,END)
    time.delete(0,END)
    content.delete(0,END)
    
#Delete information
def remove():
    selects = my_tree.selection()
    if selects:
        res = messagebox.askyesno('Remove', 'Do you want remove this information?')
        if res == True:
            for i in selects:
                my_tree.delete(i)  
        else:
            pass
    else:
        messagebox.showinfo("Error", "Please select the information you want to delete!")
    clear_entry()
    print(data)
    
# select    
def select_update():
    clear_entry()
    selected = my_tree.focus()
    values = my_tree.item(selected, "values")
    classes.insert(0, values[0])
    subject.insert(0, values[1])
    date.insert(0, values[2])
    time.insert(0, values[3])
    content.insert(0, values[4])
    
#Update information    
def update_info():
    selects = my_tree.selection()
    if selects:
        res = messagebox.askyesno('Update Information', 'Do you want update this information?')
        if res == True:
            selected = my_tree.focus()
            my_tree.item(selected, text="", values=(classes.get(), subject.get(), date.get(),time.get(), content.get()))
            
        else:
            pass
    else:
        messagebox.showinfo("Error", "Please select the information you want to update!")
    clear_entry()
def search_info():
    for info in data:
        search_time = info[2]
        if search_time == search.get():
            classes_text = info[0]
            subject_text = info[1]
            date_text = info[2]         
            time_text = info[3]
            content_text = info[4]
            messagebox.showinfo("Search information", f"Lớp học: {classes_text}, Môn học: {subject_text}, Ngày dạy: {date_text}, Thời gian dạy: {time_text}, Nội dung dạy: {content_text}")
#             break
    
#Button 
add_btn = Button(root, text="Thêm", width=10, command=add_record)
add_btn.place(x=30, y=300)
remove_btn = Button(root, text="Xóa",  width=10, command=remove)
remove_btn.place(x=150, y=300)
update_btn = Button(root, text="Sửa", width=10, command=update_info)
update_btn.place(x=30, y=350)
select_update_btn = Button(root, text="Chọn sửa", width=10, command=select_update)
select_update_btn.place(x=150, y=350)
search_btn= Button(root, text="Search", width=10, command=search_info)
search_btn.place(x=30, y=400)

root.mainloop()
```
