```python
from tkinter import * 
windows = Tk()
windows.title("Simple Calculator")
windows.geometry("400x300")
windows.eval('tk::PlaceWindow . center')

num1 = Entry(windows)
num2 = Entry(windows)

result_lbl = Label(windows, text = 'Result')

def add():
    result = int(num1.get()) + int(num2.get())
    result_lbl.config(text=result) 

def minus():
    result = int(num1.get()) - int(num2.get())
    result_lbl.config(text=result) 

def multiply():
    result = int(num1.get()) * int(num2.get())
    result_lbl.config(text=result) 

def divide():
    result = int(num1.get()) / int(num2.get())
    result_lbl.config(text=result) 


add_btn = Button(windows, text = "+", width = 10, command = add)
minus_btn = Button(windows, text = "-", width = 10,command = minus)
multiply_btn = Button(windows, text = "*", width = 10, command = multiply)
divide_btn = Button(windows, text = "/", width = 10,command = divide)


num1.pack()
num2.pack()
add_btn.pack()
minus_btn.pack()
multiply_btn.pack()
divide_btn.pack()
result_lbl.pack()

windows.mainloop()
```

```python
## Project 2 My Intro
from tkinter import * 
windows = Tk()
windows.title("My Intro")
windows.geometry("400x200")

name_lbl = Label(windows, text='My name')
first_name_lbl = Label(windows, text='First name')
last_name_lbl = Label(windows, text='Last name')
first_name_ent = Entry(windows)
last_name_ent = Entry(windows)

def write_name():
    result = "My name is " + first_name_ent.get() + " " + last_name_ent.get()
    name_lbl.config(text = result)
    
submit_btn = Button(windows, text='Submit', command = write_name)

name_lbl.grid(row = 0 , column = 0 , columnspan = 2)
first_name_lbl.grid(row = 1, column = 0)
last_name_lbl.grid(row = 2, column = 0)
first_name_ent.grid(row = 1, column = 1)
last_name_ent.grid(row = 2, column = 1)
submit_btn.grid(row = 3, column = 0, columnspan = 2)

windows.mainloop()
```

```python
## Project 3 My BMI
from tkinter import * 
from tkinter import messagebox
windows = Tk()
windows.title("My BMI")
windows.geometry("400x300")

bmi_lbl = Label(windows, text='BMI Calculator')
weight_lbl = Label(windows, text='My weight(kg):')
height_lbl = Label(windows, text='My height(m):')
weight_ent = Entry(windows)
height_ent = Entry(windows)


def submit():
    result = float(weight_ent.get())**2 / float(height_ent.get())
    messagebox.showinfo("BMI result", f"Your BMI is {result}")
    
submit_btn = Button(windows, text='Submit', command = submit)


bmi_lbl.grid(row = 0 , column = 0 , columnspan = 2)
weight_lbl.grid(row = 1, column = 0)
height_lbl.grid(row = 2, column = 0)
weight_ent.grid(row = 1, column = 1)
height_ent.grid(row = 2, column = 1)
submit_btn.grid(row = 3, column = 0, columnspan = 2)

windows.mainloop()
```
