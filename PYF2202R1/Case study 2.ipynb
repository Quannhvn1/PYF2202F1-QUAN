```python
# from tkinter import * 
from tkinter import messagebox

windows = Tk()
windows.title('Log In')
windows.geometry('600x300')

website_lbl = Label(windows, text = 'Website')
email_lbl = Label(windows, text = 'Email/Username')
password_lbl = Label(windows, text = 'Password')

website_ent = Entry(windows, text = 'website')
email_ent = Entry(windows, text = 'email')
pass_ent = Entry(windows, text = 'password')

list_info = []

def clear_entry_values():
    website_ent.delete(0, END)
    email_ent.delete(0, END)
    pass_ent.delete(0, END)
    
def add_password():
    global list_info #ch nh s dng bin list info bn ngoi  s dng trong hm
    print(f"{website_ent.get()}")
    print(f"{email_ent.get()}")
    print(f"{pass_ent.get()}")
    info = (website_ent.get(), email_ent.get(), pass_ent.get())
    list_info.append(info)
    print(list_info)
    messagebox.showinfo("Add password","Password is added successfully")
    clear_entry_values()
#Nu c tn trang web ri th pass qua.   
    
def search_info():
    print(f"looking on website: {website_ent.get()}")
    for info in list_info:
        web_name = info[0]
        if web_name == website_ent.get():
            username_text = info[1]
            password_text = info[2]
            messagebox.showinfo("Website information", f"Username:{username_text}, Password:{password_text}")
            break
            
add_btn = Button(windows, text = 'Add', command = add_password)
search_btn = Button(windows, text = 'Search', command = search_info)
generate_password_btn = Button(windows, text = 'Generate Password')


website_lbl.place(x = 20 , y = 50 )
email_lbl.place(x = 20 , y = 75)
password_lbl.place(x = 20 , y = 100)

website_ent.place(x = 120, y = 50, width = 200 )
email_ent.place(x = 120, y = 75, width = 200 )
pass_ent.place(x = 120 , y = 100, width = 200 )

add_btn.place(x = 100 , y = 120, width = 250 )
search_btn.place(x = 420 , y = 50 , height = 30)
generate_password_btn.place(x = 420 , y = 90 , height = 30)

windows.mainloop()
```
